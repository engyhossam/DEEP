```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))

def tanh(x):
    return math.tanh(x)
x = [1, 2, 3]
Wf = 0.5
Uf = 0.1
bf = 0

Wi = 0.6
Ui = 0.2
bi = 0

Wc = 0.7
Uc = 0.3
bc = 0

Wo = 0.8
Uo = 0.4
bo = 0

Wy = 4
by = 0

h = 0
c = 0

for t in range(len(x)):
    xt = x[t]

    forget_value = Wf * xt + Uf * h + bf
    f = sigmoid(forget_value)

    input_value = Wi * xt + Ui * h + bi
    i = sigmoid(input_value)

    candidate_value = Wc * xt + Uc * h + bc
    c_bar = tanh(candidate_value)

    c_new = f * c + i * c_bar

    output_value = Wo * xt + Uo * h + bo
    o = sigmoid(output_value)

    h_new = o * tanh(c_new)

    print("Time step", t + 1)
    print("x =", xt)
    print("forget gate =", round(f, 3))
    print("input gate =", round(i, 3))
    print("candidate cell state =", round(c_bar, 3))
    print("new cell state =", round(c_new, 3))
    print("output gate =", round(o, 3))
    print("new hidden state =", round(h_new, 3))
    print()

    h = h_new
    c = c_new

y_hat = Wy * h + by

print("Final hidden state =", round(h, 3))
print("Final cell state =", round(c, 3))
print("Predicted next value =", round(y_hat, 3))
```
